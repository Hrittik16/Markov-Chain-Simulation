# What is a stochastic process?
A stochastic process is a collection of random variables: ${X_{t},t \in I}$, where $X_{t}$ is the set of random variables at time $t$, and $I$ is the index set of the process.
Discrete time stochastic processes are sequences of random variables.

# Simulation of Gambler's Ruin
Gambler's ruin is an example of a stochastic process.
A gambler starts off with \\$50. They can only bet in increments of \\$1. They can only win or lose \\$1 per bet. They will keep gambling until they either lose all their money (leave with \\$0) or if they win \\$100. We can formalize this with the following notation:\
$\begin{equation}
    X_{g} = 
        \begin{cases}
            \$+1, & \text{with a probability of 50%}\\
            \$-1, & \text{with a probability of 50%}\\
        \end{cases}
\end{equation}$\
** Wherre: $X_g$ is the gambling outcome in dollars. **

```python
def gamblers_ruin():
    gambling_money = 50
    gambling_goal = 100
    gambling_simulations = []

    while gambling_money in range(1, gambling_goal):
        bet_size = 1
        w_or_l = random.randrange(-1, 2, step=2)
        gambling_money += bet_size * w_or_l
        gambling_simulations.append(gambling_money)
    return gambling_simulations
```

```python
plt.plot(gamblers_ruin())
plt.yticks(np.arange(-20,120,10))
plt.axhline(y=0, color='r', linestyle='-')
plt.axhline(y=100, color='black', linestyle='-')
plt.xlabel('Number of bets')
plt.ylabel('Winnings')
plt.title('Gambling Simulation')
```

```python
def prob_of_ruin(gambling_goal, initial_gambling_money):
    return (gambling_goal - initial_gambling_money)/gambling_goal
```

```python
prob_of_ruin(100, 50)
```

```python
# Simulating the probability of ruin
sim_list = []

while len(sim_list) < 1000:
    sim_list.append(gamblers_ruin()[-1])

print(np.mean(sim_list))
plt.hist(sim_list)
```

# Markov Chains
A markov chain is a type of stochastic process. 
A markov chain is a collection of random variales $(X_{t})$ where the future states $(j)$ only depend on the current state $(i)$. Markov chains can be either discrete or continuous.

For a markov chain transition matrix (denoted as P):
- Each row must add to one, where: $\sum_{j} P_{ij} = 1$.
- The probabilities must be non-negative where: $P_{ij} \geq 0 \ \ \forall \ \ i,j$

# Markov Chain Simulation
A markov chain can be simulated from an initial distribution/state and transition matrix. 
For eg. Let's imagine a global travel route simulation. 
Our initial state is New York City. From the initial state we can travel to: Paris, Cairo, Seoul or even within New York City.
The transition matrix contains the one step transition probabilities of moving from state to state.

```python
mc_example = {'NYC': [.25, 0, .75, 1],
              'Paris': [.25, .25, 0, 0],
              'Cairo': [.25, .25, .25, 0],
              'Seoul': [.25, .5, 0, 0]}
mc = pd.DataFrame(data=mc_example, index=['NYC', 'Paris', 'Cairo', 'Seoul'])
```

## We can formalize the movement for the Markov Chain at the initial starting point (New York City) with the following notation:
$
    P(X_{1} \ \ = \ \ New York \ \  | \ \  X_{0} \ \ = \ \ New York) \ \ = \ \ P(X_{1} \ \ = \ \ Paris \ \  | \ \  X_{0} \ \ = \ \ New York) \ \ = \ \ P(X_{1} \ \ = \ \ Cairo \ \  | \ \  X_{0} \ \ = \ \ New York)$
    $ = P(X_{1} \ \ = \ \ Seoul \ \  | \ \  X_{0} \ \ = \ \ New York) \ \ = \ \ 25% $

```python
travel_sim = []
travel_sim.append(mc.iloc[0].index[0])
city = np.random.choice(mc.iloc[0].index, p=mc.iloc[0])
travel_sim.append(city)

while len(travel_sim) < 25:
    city = np.random.choice(mc.iloc[mc.index.get_loc(city)].index, p=mc.iloc[mc.index.get_loc(city)])
    travel_sim.append(city)
```
